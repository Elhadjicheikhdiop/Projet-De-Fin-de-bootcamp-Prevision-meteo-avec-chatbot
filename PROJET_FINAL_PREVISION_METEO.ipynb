```python
import streamlit as st
import speech_recognition as sr
import requests
import json
import pyttsx3
import pickle
from gtts import gTTS
from io import BytesIO

# Chargement du modle XGBoost enregistr dans un fichier pickle
with open('xgb.pkl', 'rb') as f:
    model = pickle.load(f)

# Obtention de la ville  partir de laquelle l'utilisateur veut obtenir la mto
# Message d'introduction
st.title("Chatbot mto")
st.write("Bienvenue sur le chatbot mto ! Vous pouvez parler ou taper le nom de votre ville ou votre code postal pour connatre les prvisions mtorologiques.")

# Configuration de la reconnaissance vocale
r = sr.Recognizer()
with sr.Microphone() as source:
    st.write("Dites le nom de la ville")
    audio = r.listen(source)

# Obtention de la ville avec la reconnaissance vocale
try:
    city = r.recognize_google(audio, language='fr-FR')
    st.write(f"Vous avez demand la mto pour la ville de {city}")
except:
    st.write("Je n'ai pas compris la ville que vous avez dit.")
    city = None
st.write(f"Vous avez demand la mto pour la ville de {city}")

# Rcupration des donnes mtorologiques
if city:
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&lang=fr&appid=98e027721603ca32cead5457312b406a"
    response = requests.get(url)
    data = response.json()
    st.image(f"http://openweathermap.org/img/wn/{data['weather'][0]['icon']}.png")
    st.write(f"Conditions mtorologiques: {data['weather'][0]['description']}")
    st.write(f"Temprature: {round(data['main']['temp'] - 273.15)}C")
    st.write(f"Humidit: {data['main']['humidity']}%")
    st.write(f"Vitesse du vent: {data['wind']['speed']} m/s")

    # Fonction pour obtenir les prvisions mtorologiques  partir d'une API
def get_weather(city):
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&lang=fr&appid=98e027721603ca32cead5457312b406a"
    response = requests.get(url)
    data1 = response.json()
    return data1
    
    # Fonction pour convertir du texte en parole
def text_to_speech(text):
    with BytesIO() as file:
        gTTS(text=text, lang='fr').write_to_fp(file)
        file.seek(0)
        return file.read()

city = st.text_input("Ville ")
submit_button = st.button("Obtenir la prvision mtorologique")
        
if submit_button:
    data1 = get_weather(city)
    
    # Affichage de la rponse du chatbot et des prvisions mtorologiques
    st.write(f"Conditions mtorologiques: {data1['weather'][0]['description']}")
    st.write(f"Temprature: {round(data1['main']['temp'] - 273.15)}C")
    st.write(f"Humidit: {data1['main']['humidity']}%")
    st.write(f"Vitesse du vent: {data1['wind']['speed']} m/s")
    
    
    # Synthse vocale de la rponse du chatbot
    audio_file = text_to_speech(f"La prvision mtorologique pour {data1['name']} est {data1['weather'][0]['description']} avec une temprature de {round(data1['main']['temp'] - 273.15)} degrs Celsius, une humidit de {data1['main']['humidity']} pour cent et une vitesse de vent de {data1['wind']['speed']} mtres par seconde.")
    st.audio(audio_file, format='audio/mp3')

# Bouton pour effacer l'historique des discussions et recommencer une nouvelle conversation
if st.button("Effacer l'historique des discussions"):
    st.session_state.history = []

# Affichage de l'historique des discussions
if 'history' not in st.session_state:
    st.session_state.history = []
for message in st.session_state.history:
    st.write(message)
    audio_file = text_to_speech(message)
    st.audio(audio_file, format='audio/mp3')

# Ajout de nouveaux messages  l'historique des discussions
if submit_button:
    st.session_state.history.append(f"Prvision mtorologique pour {data1['name']}: {data1['weather'][0]['description']}, temprature de {round(data1['main']['temp'] - 273.15)}C, humidit de {data1['main']['humidity']}%, vitesse de vent de {data1['wind']['speed']} m/s.")
```
